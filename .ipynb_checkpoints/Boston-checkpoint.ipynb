### Cecha CHAS

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
df = pd.DataFrame(scaled_data)
df.columns = data.columns
chas_data = df.loc[df['CHAS'] > 0]
no_chas_data = df.loc[df['CHAS'] < 0]
chas_pairs = pd.DataFrame(columns=data.columns)
for index, ch_data in chas_data.iterrows():
    for index2, no_ch_data in no_chas_data.iterrows():
        if abs(ch_data['CRIM'] - no_ch_data['CRIM']) < 0.3 and \
        abs(ch_data['ZN'] - no_ch_data['ZN']) < 0.3 and \
        abs(ch_data['INDUS'] - no_ch_data['INDUS']) < 0.3 and \
        abs(ch_data['NOX'] - no_ch_data['NOX']) < 0.3 and \
        abs(ch_data['RM'] - no_ch_data['RM']) < 0.3 and \
        abs(ch_data['AGE'] - no_ch_data['AGE']) < 0.3 and \
        abs(ch_data['DIS'] - no_ch_data['DIS']) < 0.3 and \
        abs(ch_data['RAD'] - no_ch_data['RAD']) < 0.3 and \
        abs(ch_data['TAX'] - no_ch_data['TAX']) < 0.3 and \
        abs(ch_data['PTRATIO'] - no_ch_data['PTRATIO']) < 0.3 and \
        abs(ch_data['B'] - no_ch_data['B']) < 0.3 and \
        abs(ch_data['LSTAT'] - no_ch_data['LSTAT']) < 0.3:
            chas_pairs = chas_pairs.append(data.iloc[index])
            chas_pairs = chas_pairs.append(data.iloc[index2])
chas_pairs.head(chas_pairs.shape[0])
```

### Data split into train and test batches with [train_test_split](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) method

```python
features = boston.data 
prices = boston.target 
   

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(features, prices, test_size = 0.1, 
                                                                random_state = 1) 
```

```python
print("Training dataset:")   
print("boston_train_data: ", xtrain.shape)
print("boston_train_target", ytrain.shape) 
```

```python
print("Testing dataset:")
print("boston_test_data: ", xtest.shape) 
print("boston_test_target: ", ytest.shape) 
```

## Learning
### Training [LinearRegression](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) model

```python
from sklearn.linear_model import LinearRegression 
lin_reg = LinearRegression() 
lin_reg.fit(xtrain, ytrain) 
```

```python
lin_pred = lin_reg.predict(xtest) 
```

```python
plt.scatter(ytest, lin_pred, c = 'blue')
plt.grid()
plt.xlabel("Actual price in 1000$") 
plt.ylabel("Predicted price") 
plt.title("Actual price vs predicted price : Linear Regression") 
plt.show() 
```

### Model evaluation 


```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
mse = mean_squared_error(ytest, lin_pred) 
print("Mean Square Error : %.2f " % mse)

vs = r2_score(ytest, lin_pred)
print('Variance score: %.2f' % vs)
```

### Model evaluation with [cross-validation](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html)

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(LinearRegression(), boston['data'], boston['target'], cv=4)
print(scores)
```

```python
from sklearn.linear_model import SGDRegressor
sgd_regression = SGDRegressor(learning_rate='constant',eta0=0.01,penalty=None, max_iter=1000)
sgd_regression.fit(xtrain, ytrain)
sgd_pred = sgd_regression.predict(xtest) 
```

#### SGD  Regressor early stopping

```python
from sklearn.base import clone 

sgd_regression = SGDRegressor(learning_rate='constant', warm_start=True, penalty=None, eta0=0.0005)

minimum_val_err = float("inf")
best_epoch =  None
best_model =  None
for epoch in range(1000):
    sgd_regression.fit(xtrain, ytrain)
    y_val_predict = sgd_regression.predict(xtest)
    val_error = mean_squared_error(ytest, y_val_predict)
    if val_error < minimum_val_err:
        best_epoch =epoch
        best_model = clone(sgd_regression)
        
best_model.fit(xtrain, ytrain)
```

```python
sgd_es_pred = best_model.predict(xtest)
```

```python
plt.scatter(ytest, sgd_es_pred, c = 'blue')
plt.grid()
plt.xlabel("Actual price in 1000$") 
plt.ylabel("Predicted price") 
plt.title("Actual price vs predicted price: SGD Regression early stopping") 
plt.show() 
mse = mean_squared_error(ytest, sgd_es_pred) 
print("Mean Square Error : %.2f " % mse)
vs = r2_score(ytest, sgd_es_pred)
print('Variance score: %.2f' % vs)
```

```python
from sklearn.neural_network import MLPRegressor

neural_network = MLPRegressor(max_iter=100000,solver='adam', random_state=1)
neural_network.fit(xtrain, ytrain)
nn_pred = neural_network.predict(xtest)
plt.scatter(ytest, nn_pred, c = 'blue') 
plt.xlabel("True price") 
plt.ylabel("Predicted value") 
plt.title("True value vs predicted value : MLPRegressor") 
plt.show()
mse = mean_squared_error(ytest, nn_pred) 
print("Mean Square Error : %.2f " % mse)
vs = r2_score(ytest, nn_pred)
print('Variance score: %.2f' % vs)
```
